1. What is hypothesis testing in statistics?
    - Hypothesis testing is a formal statistical procedure used to determine whether a claim or assumption about a population is supported by sample data. It functions as a methodical framework for decision-making, allowing researchers to distinguish between genuine effects and results that likely occurred by random chance.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
    - The Null Hypothesis (H0)

      The Null Hypothesis represents the "default" position. It assumes that any observed difference or effect in your data is due to random chance rather than a specific cause.
      Typical phrasing: "There is no difference," "The treatment has no effect," or "The two groups are equal."

      The Alternative Hypothesis (Ha)
      The Alternative Hypothesis is the claim you are actually trying to prove. It proposes that the observed data shows a real effect, relationship, or difference.
      Typical phrasing: "There is a significant difference," "The new drug is more effective," or "The mean is greater than zero."
3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
    - The significance level, denoted by the letter alpha (α), is the threshold you set at the start of an experiment to define how much evidence you need before you are willing to reject the null hypothesis
    Its Role in the Decision Process
    The significance level acts as the "finish line" for your data. Once you calculate your p-value (the probability that your results happened by pure chance), you compare it to to make a final decision:
    If P<= α: The result is statistically significant. The data is "too rare" to have happened by chance under the null hypothesis. You reject the null hypothesis and support the alternative.
    If P> α: The result is not statistically significant. The data is consistent with random fluctuation. You fail to reject the null hypothesis.

4.	What are Type I and Type II errors? Give examples of each.
    - Type I Error (False Positive)
     A Type I error occurs when you reject the null hypothesis even though it is actually true. This leads you to conclude that a significant result or relationship exists when the findings were actually due to random chance
     Example (Legal): In a criminal trial, the null hypothesis is that the defendant is innocent. A Type I error occurs if an innocent person is convicted.
     Type II Error (False Negative)
     A Type II error occurs when you fail to reject the null hypothesis even though it is actually false. This means you miss a real effect or difference that was present in the population.
     Example (Legal): In a trial, a Type II error occurs if a guilty person is acquitted (found not guilty).

5.	What is the difference between a Z-test and a T-test? Explain when to use each.
    - The primary difference between a Z-test and a T-test lies in whether the   population standard deviation (σ) is known and the size of the sample being analyzed.
     Use a Z-Test when:
     The population standard deviation (σ) is known.
     The sample size is large (typically n>=30).
     You are comparing a sample mean to a population mean or testing population proportions.
     Use a T-Test when:
     The population standard deviation (σ) is unknown.
     The sample size is small (typically n<30).

```python
#7.	Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data provided
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
null_mean = 50.0

# Perform one-sample Z-test
z_stat, p_value = ztest(sample_data, value=null_mean)

print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the Null Hypothesis (Statistically Significant)")
else:
    print("Result: Fail to Reject the Null Hypothesis (Not Statistically Significant)")
```

```python
#8.	Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Simulate data from a Normal Distribution (mean=50, std_dev=5, n=100)
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)

# 2. Calculate mean and 95% Confidence Interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Calculate the Standard Error and the Margin of Error
standard_error = sample_std / np.sqrt(n)
z_critical = stats.norm.ppf(0.975)  # 1.96 for 95% CI
margin_of_error = z_critical * standard_error

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")


plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='lightgreen', edgecolor='black', alpha=0.7, label='Sample Data')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.axvspan(confidence_interval[0], confidence_interval[1], color='red', alpha=0.15, label='95% CI')

plt.title('Normal Distribution Simulation & Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#9.	Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):

    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores: (x - mean) / std_dev
    z_scores = (data - mean) / std_dev


    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=15, color='plum', edgecolor='black', alpha=0.8)


    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axvline(1, color='blue', linestyle=':', label='+1 SD')
    plt.axvline(-1, color='blue', linestyle=':', label='-1 SD')

    plt.title('Standardized Data (Z-Scores)')
    plt.xlabel('Z-score (Distance from Mean in SD)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    return z_scores


data_sample = np.random.normal(100, 15, 500)
z_results = plot_z_scores(data_sample)
```
